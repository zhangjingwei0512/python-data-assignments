```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://finance.yahoo.com/quote/%5EHSI/history?p=%5EHSI'
r=requests.get(url)
r.ecoding='utf-8'
html_str=r.text
data=BeautifulSoup(html_str,"html.parser")
a=data.find_all('td')
```

```python
list=[]
for i in range(0,len(a)):
    list.append(a[i].text.strip())
list
```

```python
for i in range(0,len(list)):
    if i%7!=0:
        list[i]=list[i].replace(',','')
list
```

```python
import csv
with open('HANG SENG INDEX.csv','w',newline='')as f:
    writer=csv.writer(f)
    header=['date','open','high','low','close','Adj Close','volume']
    writer.writerow(header)
    for i in range(0,100):
        writer.writerow([list[i*7],list[i*7+1],list[i*7+2],list[i*7+3],list[i*7+4],list[i*7+5],list[i*7+6]])
```

```python
import pandas as pd
df=pd.read_csv('HANG SENG INDEX.csv')
df
```

```python
df.describe()
```

```python
df['difference']=df['close']-df['open']
df['rate']=df['difference']/df['open']
del df['Adj Close']
df
```

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df['datetime'] = df['date'].apply(parse_datetime)
df
```

```python
df.set_index('datetime').resample('M').aggregate('rate').plot()
```

```python
#news point 1:
#In these four months, the rate of the index change has maintained in the range of (-0.03,0.03)
#This range shows the stability of the Hang Seng Index 
#But the extreme points occur in these two months, which means the market fluctuates wildly compared  with the previous months
```

```python
import seaborn as sns
sns.regplot(df['high'], df['low'])
ax = sns.scatterplot(x="high", y="low",hue='difference',data=df)
```

```python
import seaborn as sns
sns.regplot(df['open'], df['close'])
ax = sns.scatterplot(x="open", y="close",hue='difference',data=df)
```

```python
#news point 2:
#The performance of Hang Seng Index conforms to market rules.
#The low price tend to be higher if the high price is higher
#The close price tend to be higher if the open price is higher
#Most of the points are distributed near the line with only a few points represent of the difference.
#The value of the close price could be approximately calculated through a linear equation 
```
